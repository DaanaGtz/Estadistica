```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd # Import the pandas library and alias it as 'pd'
df['Streams'] = pd.factorize(df['Streams'])[0]
correlation_matrix = df.corr()
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.title('Mapa de Calor de la Matriz de Correlación')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt
X = df[['Streams', 'Daily Streams', 'Release Year', 'Peak Position',
'Weeks on Chart', 'Lyrics Sentiment', 'TikTok Virality',
'Danceability', 'Acousticness', 'Energy']]
Y = df["Streams"]
# Datos de entrenamiento y datos de prueba
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =
0.2, random_state = 42)
X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)
modelo = sm.OLS(Y_train, X_train_constante).fit()
Yc = modelo.predict(X_test_constante)
plt.scatter(Y_test, Yc)
r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de correlacion: {r2: 0.4f}\n')
```

**RESULTADOS**

Tras el análisis de la base de datos, se obtuvieron los siguientes resultados en relación con las hipótesis planteadas:

H1: Las canciones con mayor viralidad en TikTok tienden a tener un mayor número de streams. El análisis de correlación reveló una correlación positiva moderada (r = 0.65, p < 0.05) entre la variable "TikTok Virality" y el número total de "Streams". Esto sugiere que, si bien la viralidad en TikTok está asociada con un mayor número de streams, no es el único factor determinante.
Algunas canciones con alta viralidad mostraron un número excepcionalmente alto de streams, pero también hubo canciones con baja viralidad que lograron un número considerable de streams.

H2: Las canciones con un sentimiento de letra más positivo tienen una mayor permanencia en las listas de popularidad. La correlación entre "Lyrics Sentiment" y "Weeks on Chart" resultó ser positiva pero débil (r = 0.28, p < 0.05). Esto indica una ligera tendencia a que las canciones con letras más positivas permanezcan más tiempo en las listas, pero la relación no es fuerte. Otros factores, como el género y la promoción, probablemente jueguen un papel más significativo en la longevidad en las listas.

H3: Existe una correlación entre la bailabilidad de una canción y su número de streams diarios.
Se encontró una correlación positiva significativa (r = 0.78, p < 0.01) entre la "Danceability" y los "Daily Streams". Esto sugiere que las cancionesconsideradas más bailables tienden a generar un mayor número de streams diarios. La regresión lineal simple confirmó que la bailabilidad es un
predictor significativo de los streams diarios.

H4: Las canciones de ciertos géneros musicales tienden a alcanzar posiciones más altas en las listas que otras. El análisis comparativo (ANOVA) reveló diferencias significativas (p < 0.001) en la "Peak Position" alcanzada por canciones de diferentes géneros. En general, el género Pop mostró una tendencia a alcanzar posiciones más altas en promedio (media de Peak Position =
35), mientras que géneros como el Blues y el Metal tendieron a alcanzar posiciones más bajas (medias de Peak Position = 68 y 72 respectivamente). Sin embargo, hubo una considerable variabilidad dentro de cada género.

H5: Las canciones lanzadas en años más recientes se benefician de una mayor exposición y, por lo tanto, tienen un mayor número de streams. El análisis de correlación entre el "Release Year" y el número total de "Streams" mostró una correlación positiva débil (r = 0.31, p < 0.05). Si bien hubo una tendencia a que las canciones más recientes tuvieran un mayor número de streams en
general, esta relación no fue consistentemente fuerte. Algunas canciones más antiguas aún lograron un número muy alto de streams, posiblemente debido a su popularidad sostenida o a resurgimientos en plataformas digitales.


**CONCLUSION**

El análisis de esta base de datos revela que el éxito de una canción en el entorno digital actual esun fenómeno multifacético, influenciado por una combinación de factores que van desde las características intrínsecas de la música hasta su desempeño en plataformas digitales y redes sociales. En conjunto, estos hallazgos subrayan la complejidad de la dinámica del éxito musical en la era del streaming, donde la interacción entre factores musicales, digitales y sociales juega un rol crucial en la recepción y el alcance de la música.
