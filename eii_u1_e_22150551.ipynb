```python
#Examen Daana Paola Gutierrez Garcia
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/DaanaGtz/Estadistica/refs/heads/main/mental_health.csv")
df

#a) hip√≥tesis de causalidad: Redacta una hip√≥tesis sobre la causalidad entre las dos variables, a su vez, establece la variable dependiente y la variable independiente.
#Se observa una diferencia significativa en la tasa de diagn√≥sticos de ansiedad entre individuos de 25 a 34 a√±os que experimentan altos niveles de estr√©s laboral y aquellos que no.

X = df["Year"]
Y = df["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"]

#b)Recta de regresi√≥n ajustada: Realiza los c√°lculos necesarios para encontrar la recta de regresi√≥n ajustada. Incluye los procedimientos que te llevaron a la recta de regresi√≥n (Colab)
import statsmodels.api as sm
X_constante = sm.add_constant(X)
modelo = sm.OLS(Y, X_constante).fit()
b0, b1 = modelo.params
Fun = b0 + b1 * X

print("Recta de regresi√≥n ajustada:")
print(f"Y = {b0:.2f} + {b1:.2f} * X")

#c)grafica los puntos y la recta de regresi√≥n
import matplotlib.pyplot as plt
plt.scatter(X, Y, label='Datos')
plt.plot(X, Fun, color='brown', label='Recta de regresi√≥n')
plt.xlabel('Year')
plt.ylabel('Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized')
plt.title('Regresi√≥n Lineal Simple')
plt.legend()
plt.show()

#d)Coeficiente de correlaci√≥n y determinaci√≥n: Realiza los c√°lculos necesarios para obtener la r de Pearson y el coeficiente de determinaci√≥n. Incluye dichos c√°lculos
from scipy.stats import pearsonr
r,_ = pearsonr(X, Y)
r_cuadrado = r**2
print("Coeficiente de correlaci√≥n (r):", r)
print("Coeficiente de determinaci√≥n (r^2):", r_cuadrado)

#e)Interpreta los resultados del coeficiente de correlaci√≥n y el coeficiente de determinaci√≥n
#Coeficiente de correlaci√≥n: El coeficiente de correlaci√≥n obtenido en este an√°lisis es de 0.02948647738938569. Este valor, extremadamente cercano a cero, indica una correlaci√≥n lineal pr√°cticamente inexistente entre las dos variables estudiadas. En t√©rminos estad√≠sticos, esto implica que la variaci√≥n en una de las variables no se refleja en cambios proporcionales en la otra.
#Coeficiente de determinaci√≥n:El valor obtenido para el coeficiente de determinaci√≥n, r^2, es de 0.0008694523488347536. Este resultado, extremadamente pr√≥ximo a cero, revela que la variable independiente considerada tiene una capacidad pr√°cticamente nula para explicar la variabilidad observada en la variable dependiente. En t√©rminos m√°s sencillos, esto implica que los cambios en la variable independiente no se traducen en cambios significativos en la variable dependiente, y viceversa.

#f)Calcula el intervalo de confianza del 95% para Œ≤_1 y Œ≤_0 , ¬øQu√© dice el intervalo de confianza a la posibilidad de que Œ≤_1 sea igual a cero?
nivel_de_confianza = 0.95
intervalo_de_confianza = modelo.conf_int(alpha=1-nivel_de_confianza)
print("Intervalo de confianza del 95% para Œ≤_0:", intervalo_de_confianza[0])
print("Intervalo de confianza del 95% para Œ≤_1:", intervalo_de_confianza[1])

#g) Realiza el gr√°fico de los residuales: A partir de este gr√°fico, menciona si los datos cumplen con los supuestos para la regresi√≥n: linealidad, normalidad en torno a la recta, homoscedasticidad (hay m√°s, pero con estas nos bastan)
# Gr√°fico de residuales vs. valores predichos
residuales = modelo.resid
# Gr√°fico de residuales vs. valores predichos
plt.scatter(modelo.fittedvalues, residuales)
plt.axhline(y=0, color='orange', linestyle='--')
plt.xlabel('Valores predichos')
plt.ylabel('Residuales')
plt.title('Gr√°fico de Residuales vs. Valores Predichos')
plt.show()
# Gr√°fico de residuales vs. variable independiente
plt.scatter(df['Year'], residuales)
plt.axhline(y=0, color='green', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Residuales')
plt.title('Gr√°fico de Residuales vs. Year')
plt.show()
#Normalidad en torni a la recta
import scipy.stats as stats
import matplotlib.pyplot as plt
#Histograma de los residuales
plt.hist(residuales, bins=20)
plt.xlabel('Residuales')
plt.ylabel('Frecuencia')
plt.title('Histograma de Residuales')
plt.show()
#Gr√°fico Q-Q de los residuales
stats.probplot(residuales, dist="norm", plot=plt)
plt.title('Gr√°fico Q-Q de Residuales')
plt.show()
#Homoscedasticidad
plt.scatter(modelo.fittedvalues, residuales)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Valores predichos')
plt.ylabel('Residuales')
plt.title('Gr√°fico de Residuales vs. Valores Predichos')
plt.show()
```
